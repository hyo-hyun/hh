```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_ary = te.fit(filtered_transactions).transform(filtered_transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# apriori 실행 (min_support 설정은 필요에 따라 조절)
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)

# 연관규칙 생성 (min confidence 조절 가능)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# 필요한 컬럼만 정리해서 출력 (antecedents, consequents, support, confidence, lift)
rules_simple = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].copy()
rules_simple['antecedents'] = rules_simple['antecedents'].apply(lambda x: ', '.join(list(x)))
rules_simple['consequents'] = rules_simple['consequents'].apply(lambda x: ', '.join(list(x)))
rules_simple_sorted = rules_simple.sort_values(by='lift', ascending=False)
print(rules_simple_sorted.head(20))
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# transactions: 이미 만들어진 거래 데이터 리스트 (여기선 리스트 of 리스트 형태)
# 예) transactions = [['강구안', '동피랑'], ['고양이학교'], ['동피랑', '디피랑', '강구안'], ...]

# 1. 거래 데이터 one-hot 인코딩
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# 2. 자주 등장하는 관광지 집합(frequent itemsets) 구하기
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)  # support 기준은 필요에 따라 조정

# 3. 연관규칙 생성
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)

# 4. 결과 중 필요한 컬럼만 추출 + frozenset → 문자열 변환
rules_simple = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].copy()
rules_simple['antecedents'] = rules_simple['antecedents'].apply(lambda x: ', '.join(list(x)))
rules_simple['consequents'] = rules_simple['consequents'].apply(lambda x: ', '.join(list(x)))

#출력
rules_simple_sorted = rules_simple.sort_values(by='lift', ascending=False)
print(rules_simple_sorted.head(20))
```
