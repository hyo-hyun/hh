```python
# Define feature groups
id_features = ['row_id','sequence_id', 'subject'] # categorical 
train_only_features = ['sequence_type', 'orientation', 'behavior', 'phase', 'gesture'] # categorical
sequence_features = ['sequence_counter']
accel_features = ['acc_x', 'acc_y', 'acc_z']
rot_features = ['rot_w', 'rot_x', 'rot_y', 'rot_z']
thm_features = ['thm_1', 'thm_2', 'thm_3', 'thm_4', 'thm_5']
tof_features = [f"tof_{i}_v{j}" for i in range(1, 6) for j in range(64)]
demo_features =  ['adult_child', 'age', 'sex', 'handedness', 'height_cm','shoulder_to_wrist_cm', 'elbow_to_wrist_cm']

# Define feature groups
feature_groups = {
    "id_features": id_features,
    "train_only_features": train_only_features,
    "sequence_feature": sequence_features,
    "accel_features": accel_features,
    "rot_features": rot_features,
    "thm_features": thm_features,
    "tof_features": tof_features,
    "demo_features": demo_features
}

# Print the number of features in each group
for group_name, features in feature_groups.items():
    print(f"{group_name}: {len(features)}")

print()
# Calculate total feature count
total_features = sum(len(features) for features in feature_groups.values())

print(f"Total number of features: {total_features}")
```
